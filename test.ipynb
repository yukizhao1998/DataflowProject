```python
from datacenter import Datacenter
from config import Config
```

```python
def generate_datacenters(datacenter_num):
    datacenter_list=[]
    for i in range(datacenter_num):
        datacenter_list.append(Datacenter(i))
    for i in range(datacenter_num):
        datacenter_list[i].set_datacenters(datacenter_list)
    return datacenter_list
```

```python
# initialize configurations
conf=Config()
# initialize datacenters, each element is a Datacenter type item
datacenters=generate_datacenters(conf.datacenter_num)
```

```python
cost_pi = 0
for i in range(conf.datacenter_num):
    cost_pi += datacenters[i].test("P_I")
print("Perfect indicator cost:", cost_pi)
```

```python
cost_epi = 0
for i in range(conf.datacenter_num):
    cost_epi += datacenters[i].test("epi")
print("epi cost:", cost_epi)
```

```python
cost_cpi = 0
for i in range(conf.datacenter_num):
    cost_cpi += datacenters[i].test("cpi")
print("cpi cost:", cost_cpi)
```
