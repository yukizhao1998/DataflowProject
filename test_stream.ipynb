```python
from datacenter import Datacenter
from config import Config
import random
```

```python
def generate_datacenters(datacenter_num):
    datacenter_list=[]
    for i in range(datacenter_num):
        datacenter_list.append(Datacenter(i))
    for i in range(datacenter_num):
        datacenter_list[i].set_datacenters(datacenter_list)
    return datacenter_list
```

```python
# initialize configurations
conf=Config()
# initialize datacenters, each element is a Datacenter type item
datacenters=generate_datacenters(conf.datacenter_num)
queries=[]
with open(conf.load_query_path, 'r', encoding='utf-8') as file:
    while True:
        line = file.readline()
        if line:
            query = []
            key = line.strip("\n")
            datacenter_id = random.randint(0, conf.datacenter_num - 1)
            query.append(key)
            query.append(datacenter_id)
            queries.append(query)
        else:
            break
```

```python
def test(method, queries):
    acc_cost = 0
    miss_cost = 0
    total_cost = 0
    count = 0
    query_num = len(queries)
    print("Test:", method)
    if (method == 'P_I'):
        for query in queries:
            if count % 10000 == 0:
                print("%d / %d completed, with cost %d, %d, %d."%(count, query_num, acc_cost, miss_cost, total_cost))
            count += 1
            a, m = datacenters[query[1]].perfect_indicator(query[0])
            acc_cost += a
            miss_cost += m
            total_cost = acc_cost + miss_cost
    elif (method == 'epi'):
        for query in queries:
            if count % 10000 == 0:
                print("%d / %d completed, with cost %d, %d, %d."%(count, query_num, acc_cost, miss_cost, total_cost))
            count += 1
            a, m = datacenters[query[1]].epi(query[0])
            acc_cost += a
            miss_cost += m
            total_cost = acc_cost + miss_cost
    elif (method == 'cpi'):
        for query in queries:
            if count % 10000 == 0:
                print("%d / %d completed, with cost %d, %d, %d."%(count, query_num, acc_cost, miss_cost, total_cost))
            count += 1
            a, m = datacenters[query[1]].cpi(query[0])
            acc_cost += a
            miss_cost += m
            total_cost = acc_cost + miss_cost
    elif (method == 'DSpot'):
        Datacenter.query_cnt = [0] * 17
        Datacenter.miss_cnt = [0] * 17
        Datacenter.missind_ratio = [0] * 17
        for query in queries:
            if count % 10000 == 0:
                print("%d / %d completed, with cost %d, %d, %d."%(count, query_num, acc_cost, miss_cost, total_cost))
            count += 1
            a, m = datacenters[query[1]].DSpot(query[0])
            acc_cost += a
            miss_cost += m
            total_cost = acc_cost + miss_cost
    
    print("%s test completed with cost %d, %d, %d"%(method, acc_cost, miss_cost, total_cost))
```

```python
test("DSpot", queries)
```

```python
test("cpi", queries)
```

```python
test("P_I", queries)
```
